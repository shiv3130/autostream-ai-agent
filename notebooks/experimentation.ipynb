```python
# AutoStream AI Agent – Experimentation Notebook

This notebook was used to experiment with the RAG pipeline, embeddings, and retrieval quality during development.

Goals:
- Test embedding models
- Validate document chunking
- Evaluate vector search results
- Experiment with LLM prompts

Final production code was implemented in the `src/` directory.
```

```python
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.vectorstores import FAISS
from langchain_huggingface import HuggingFaceEmbeddings
```

```python
loader = TextLoader("../data/knowledge_base.md")
documents = loader.load()

print("Number of documents:", len(documents))
documents[0]
```

```python
splitter = RecursiveCharacterTextSplitter(
    chunk_size=400,
    chunk_overlap=50
)

docs = splitter.split_documents(documents)

print("Total chunks:", len(docs))
docs[0]
```

```python
embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)

vectorstore = FAISS.from_documents(docs, embeddings)

print("Vector database created successfully")
```

```python
query = "What is the AutoStream Pro plan?"

results = vectorstore.similarity_search(query, k=3)

for r in results:
    print(r.page_content)
    print("-----")
```

```python
context = "\n".join([doc.page_content for doc in results])

prompt = f"""
Context:
{context}

Question:
What does the Pro plan include?
"""

print(prompt)
```
